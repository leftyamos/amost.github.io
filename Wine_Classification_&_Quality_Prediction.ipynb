## Data Preprocessing

```python
x = wine.drop('label',axis=1)

y = wine['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=101)
```

```python
# encode features
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

## Model: Binary Classification


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
```

#### Model 1

```python
model = Sequential()

model.add(Dense(12, activation='relu', input_shape=(X_train.shape[1],)))

model.add(Dense(8, activation='relu'))

model.add(Dense(1,activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, 
          batch_size=1, 
          epochs=20, verbose=1)
```

```python
model_loss = pd.DataFrame(model.history.history)

model_loss.plot()
```

#### Predict & Evaluate Model 1

```python
# prediction
pred = model.predict_classes(X_test)
```

```python
# evaluation
loss, acc = model.evaluate(X_test, y_test, verbose=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

#### Model 2

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=0, patience=25)
```

```python
model = Sequential()

model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.5))

model.add(Dense(16, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(8, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, 
          batch_size=20, 
          epochs=100, verbose=1,
          validation_data=(X_test, y_test),
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)

model_loss.plot()
```

#### Predict & Evaluate Model 2

```python
# prediction
pred = model.predict_classes(X_test)
```

```python
# evaluation
loss, acc = model.evaluate(X_test, y_test, verbose=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, pred))
```

## Model: Scalar Regression using K-fold cross-validation

```python
# Data Preprocessing
x = wine.drop('quality',axis=1)

y = wine['quality']
```

```python
# scale
X = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=101)

# Build Model
for train, test in kfold.split(X, y):
    model = Sequential()
    model.add(Dense(64, input_dim=12, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    model.fit(X[train], y[train], epochs=10, verbose=1)
```

```python
# evaluation
mse_value, mae_value = model.evaluate(X[test], y[test], verbose=1)
```

```python
# prediction
pred = model.predict(X[test])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
mean_absolute_error(y[test], pred)
```

```python
mean_squared_error(y[test], pred)
```

```python
explained_variance_score(y[test], pred)
```

```python
# Our predictions
plt.scatter(y[test],pred)

# Perfect predictions
plt.plot(y[test],y[test],'r')
```

## Model: Mutli-label Classification using K-fold cross-validation

```python
# Data Preprocessing
x = wine.drop('quality',axis=1)

y = wine['quality']
```

```python
# scale
X = scaler.fit_transform(x)
```

```python
y.value_counts()
```

```python
# set labels starting from zero
y = y.map({3:0,4:1,5:2,6:3,7:4,8:5,9:6})
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=101)

# Build Model
for train, test in kfold.split(X, y):
    model = Sequential()
    model.add(Dense(12, input_dim=12, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(7,activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train], y[train], epochs=20, verbose=1)
```

```python
# evaluation
loss, acc = model.evaluate(X[test], y[test], verbose=1)
```

```python
pred = model.predict_classes(X[test])
```

```python
print(classification_report(y[test], pred))
```

```python
print(confusion_matrix(y[test], pred))
```

##Model: Random Forest

```python
x = wine.drop('quality',axis=1)

y = wine['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=101)
```

```python
X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=4)

rfc.fit(X_train, y_train)
```

```python
# Mean Accuracy of Random Forest classifier on training set
rfc.score(X_train, y_train)
```

```python
# Mean Accuracy of Random Forest classifier on testing set
rfc.score(X_test, y_test)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
print(confusion_matrix(y_test, rfc_pred))
```
